```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
import pandas

file_string = "fourCircles.txt"
train_data = pandas.read_csv(file_string, sep=" ", header=None)
train_data.columns = ["x", "y"]

print(train_data["x"])

plt.scatter(train_data["x"], train_data["y"])
plt.show()
```

```python
# Identify a seed for consistent results
seed = 4
# Sample K Random Points
train_data_random_point = train_data.sample(n = 4, random_state = seed)

plt.scatter(train_data_random_point["x"], train_data_random_point["y"])
plt.show()
```
